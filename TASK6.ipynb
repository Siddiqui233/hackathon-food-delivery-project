<h1>Smart Parking Lot Management System
</h1>

```python
def smart_parking_system(capacity, vehicle_logs):
    current_parked = 0
    peak_usage = 0

    for log in vehicle_logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            if current_parked > 0:
                current_parked -= 1

        # Track peak usage
        if current_parked > peak_usage:
            peak_usage = current_parked

    # Display results
    print("Currently Parked Vehicles:", current_parked)
    print("Peak Parking Usage:", peak_usage)

    if current_parked > capacity:
        print("Parking Status: Exceeded Capacity! Alert!")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, vehicle_logs)
```

```python
def estimate_delivery_time(distance, traffic_level, weather):

    # First, calculate normal delivery time
    # Assume delivery takes 5 minutes per kilometer
    time = distance * 5

    # Now check traffic conditions
    if traffic_level.lower() == "high":
        time = time + 15      # heavy traffic adds 15 minutes
    elif traffic_level.lower() == "medium":
        time = time + 8       # medium traffic adds 8 minutes
    else:
        time = time + 0       # low traffic adds no delay

    # Now check weather conditions
    if weather.lower() == "rainy":
        time = time + 10      # rain adds 10 minutes
    elif weather.lower() == "stormy":
        time = time + 20      # storm adds 20 minutes
    else:
        time = time + 0       # clear weather adds no delay

    # Finally print the result
    print("Estimated Delivery Time:", time, "minutes")


# Given Input
estimate_delivery_time(8, "High", "Rainy")
```

```python
def monitor_classroom_resources(resource_usage):

    overused_resources = []

    # Check each resource and its usage time
    for resource, hours in resource_usage.items():

        # If usage is more than 8 hours, mark it as overused
        if hours > 8:
            overused_resources.append(resource)

    # Print results
    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_classroom_resources(usage)
```

<h1>Online Event Registration Capacity Controller
</h1>

```python
def manage_event_registration(capacity, total_registrations):

    confirmed = 0
    waitlisted = 0

    # Loop through each registration
    for i in range(total_registrations):

        # If seats are available
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # Display results
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if confirmed == capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
manage_event_registration(100, 105)
```
