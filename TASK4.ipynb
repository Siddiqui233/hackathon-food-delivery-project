<h1>Mobile Recharge Validation System</h1>

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    
    if amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans are:", valid_plans)
        return False
    
    print("Recharge successful for ₹", amount)
    return True


# Allow user to retry until valid input
while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        
        if validate_recharge(amount):
            break
        else:
            print("Please try again.\n")
    
    except ValueError:
        print("Please enter a valid number.\n")
```

<h1>Inventory Reorder Alert System
</h1>


```python
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(product, ":", quantity, "- Reorder Alert")
        else:
            print(product, ":", quantity, "- Stock OK")


# Sample inventory dictionary
inventory = {
    "Rice": 25,
    "Wheat": 10,
    "Sugar": 18,
    "Oil": 8
}

# Function call
check_inventory(inventory)
```

<h1>Student Result Processing System</h1>

```python
def process_result(marks):
    total = 0
    
    # Calculate total using loop
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Example marks list
student_marks = [60, 45, 70, 55]

result = process_result(student_marks)

print("Average Marks:", sum(student_marks)/len(student_marks))
print("Result:", result)
```

<h1>Cab Fare Estimator with Retry Option</h1>

```python
def calculate_fare(distance, is_peak_hour=False):
    base_fare = 50
    rate_per_km = 12
    
    fare = base_fare + (distance * rate_per_km)
    
    if is_peak_hour:
        fare += fare * 0.25  # Add 25% extra for peak hours
    
    return fare


# Function to get user input and retry if invalid
while True:
    try:
        distance = float(input("Enter distance in km: "))
        
        if distance <= 0:
            print("Distance must be greater than 0. Please try again.\n")
            continue
        
        peak_hour_input = input("Is it peak hour? (yes/no): ").strip().lower()
        is_peak_hour = peak_hour_input == "yes"
        
        fare = calculate_fare(distance, is_peak_hour)
        
        print(f"Total fare: ₹{fare:.2f}")
        break  # Exit the loop if fare is calculated successfully
        
    except ValueError:
        print("Invalid input. Please enter a valid number for distance.\n")
```

<h1>Employee Attendance Eligibility Checker
</h1>

```python
def check_eligibility(attendance_list):
    total_days = len(attendance_list)  # Total number of days
    present_days = attendance_list.count("P")  # Count of present days
    
    # Calculate attendance percentage
    attendance_percentage = (present_days / total_days) * 100
    
    # Return eligibility based on the percentage
    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example attendance list
attendance = ["P", "A", "P", "P", "A", "P", "P", "P", "A", "P"]

result = check_eligibility(attendance)
print(f"Attendance Percentage: {(attendance.count('P') / len(attendance)) * 100:.2f}%")
print("Eligibility Status:", result)
```

<h1>Password Strength Checker
</h1>

```python
def check_password_strength(password):
    # Rule 1: Minimum length of 8 characters
    if len(password) < 8:
        return "Password is too short, minimum length is 8."
    
    # Rule 2: Must contain at least one digit
    has_digit = False
    # Rule 3: Must contain at least one special character (@, #, $)
    has_special_char = False
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special_char = True
    
    if not has_digit:
        return "Password must contain at least one digit."
    
    if not has_special_char:
        return "Password must contain at least one special character (@, #, or $)."
    
    # If all conditions are satisfied
    return "Password is strong."

# Example usage
password = input("Enter your password: ")
strength = check_password_strength(password)
print(strength)
```
