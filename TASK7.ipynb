<h1>Smart Light Controller</h1>

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # Light is OFF by default

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")


# Sample Input
light_name = "Bedroom Light"
action = "ON"

# Creating object
light = SmartLight(light_name)

# Performing action
if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()

# Display Output
light.display_status()
```

<h1>Employee ID Card System
</h1>

```python
# Employee ID Card System

class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # Method to display ID card details
    def display_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample Input
name = "Majeed"
emp_id = "EMP102"
department = "IT"

# Creating object
employee = Employee(name, emp_id, department)

# Displaying ID Card
employee.display_id_card()
```

<h1>Mobile Contact Record
</h1>

```python
# Mobile Contact Record

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    # Method to display contact details
    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Sample Input
name = "Majeed"
phone = "9524136810"

# Creating contact object
contact1 = Contact(name, phone)

# Displaying contact information
contact1.display_contact()
```

<h1>Product Price Tag Generator</h1>

```python
# Product Price Tag Generator

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    # Method to print price tag
    def print_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))


# Sample Input
product_name = "Smart Watch"
price = 2999

# Creating product object
item = Product(product_name, price)

# Printing price tag
item.print_price_tag()
```

```python
# Movie Rating Display System

class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    # Method to display movie details
    def display_details(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


# Sample Input
movie_name = "Inception"
rating = 4.8

# Creating movie object
movie1 = Movie(movie_name, rating)

# Displaying movie details
movie1.display_details()
```

<h1>Delivery Address Manager</h1>

```python
# Delivery Address Manager

class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    # Method to print delivery details
    def print_delivery_details(self):
        print("Delivery Details")
        print("Customer Name:", self.customer_name)
        print("Address:", self.address)


# Sample Input
name = "Majeed"
address = "Hyderabad"

# Creating delivery object
order1 = Delivery(name, address)

# Printing delivery details
order1.print_delivery_details()
```
