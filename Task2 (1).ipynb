<h1>Unique Words in a Sentence</h1>

```python
# Function to count and display unique words
def unique_words_count(sentence):
    # Split the sentence into words
    words = sentence.split()
    
    # Convert the list of words into a set to remove duplicates
    unique_words = set(words)
    
    # Print the count of unique words and the unique words themselves
    print(f"Unique words count: {len(unique_words)}")
    print(f"Unique words: {unique_words}")

# Input: sentence from the user
sentence = input("Enter sentence: ")

# Call the function with the input sentence
unique_words_count(sentence)
```

<h1>Highest Salary from Employee Data
</h1>

```python
# Function to find the employee with the highest salary
def highest_salary(employees):
    # Initialize variables to store the highest salary and corresponding employee
    highest_employee = ""
    highest_salary = 0
    
    # Iterate through the dictionary to find the highest salary
    for employee, salary in employees.items():
        if salary > highest_salary:
            highest_salary = salary
            highest_employee = employee
    
    # Print the result
    print(f"Highest Salary: {highest_employee} - {highest_salary}")

# Input: Employee data in a dictionary format
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Call the function with the employee data
highest_salary(employees)
```

<h1>Find Maximum and Minimum Values</h1>

```python
# Function to find the maximum and minimum values in a list
def find_max_min(numbers):
    # Initialize max and min with the first element of the list
    max_value = min_value = numbers[0]
    
    # Loop through the list and update max and min values
    for num in numbers:
        if num > max_value:
            max_value = num
        elif num < min_value:
            min_value = num
    
    # Print the results
    print(f"List: {numbers}")
    print(f"Max: {max_value}")
    print(f"Min: {min_value}")

# Input: A list of numbers
numbers = [45, 22, 89, 10, 66]

# Call the function with the numbers list
find_max_min(numbers)
```

<h1>Count Products Above a Price Threshold</h1>

```python
# Function to count products above the price threshold
def count_products_above_threshold(prices, threshold=1000):
    count = 0  # Initialize the counter
    
    # Loop through each price in the list
    for price in prices:
        if price > threshold:
            count += 1  # Increment the counter if the price is above the threshold
    
    # Print the result
    print(f"Products above {threshold}: {count}")

# Input: List of product prices
prices = [450, 1200, 899, 1500, 300]

# Call the function with the prices list
count_products_above_threshold(prices)
```

<h1> Calculate Attendance Percentage
</h1>

```python
# Function to calculate attendance percentage
def calculate_attendance_percentage(attendance):
    # Count the number of 'P' (Present)
    present_days = attendance.count("P")
    
    # Total number of days (length of the list)
    total_days = len(attendance)
    
    # Calculate the attendance percentage
    attendance_percentage = (present_days / total_days) * 100
    
    # Print the result
    print(f"Attendance Percentage: {attendance_percentage:.1f}")

# Input: List of attendance records
attendance = ["P", "P", "A", "P", "P"]

# Call the function with the attendance list
calculate_attendance_percentage(attendance)
```

<h1>Count Character Frequency
</h1>

```python
# Function to count the frequency of each character in the string
def count_character_frequency(string):
    # Initialize an empty dictionary to store character frequencies
    char_count = {}
    
    # Loop through each character in the string
    for char in string:
        if char in char_count:
            # If the character is already in the dictionary, increment its count
            char_count[char] += 1
        else:
            # If the character is not in the dictionary, add it with count 1
            char_count[char] = 1
    
    # Print the result
    print(char_count)

# Input: String
input_string = "python"

# Call the function with the input string
count_character_frequency(input_string)
```

<h1>Check if a Key Exists in a Dictionary
</h1>

```python
# Function to check if a key exists in the dictionary
def check_key_exists(dictionary, key):
    # Check if the key exists in the dictionary using the 'in' keyword
    if key in dictionary:
        print("Employee exists")
    else:
        print("Employee does not exist")

# Example dictionary
employee_data = {
    "Mujeeb": 50000,
    "Majeed": 60000,
    "Ahmed": 55000
}

# Input: Key to check
key_to_check = input("Enter the employee name to check: ")

# Call the function with the dictionary and the key to check
check_key_exists(employee_data, key_to_check)
```

<h1>Calculate Average Marks
</h1>

```python
# Function to calculate the average marks
def calculate_average(marks):
    # Calculate the sum of all marks
    total_marks = sum(marks)
    
    # Calculate the number of marks
    num_marks = len(marks)
    
    # Calculate the average marks
    average_marks = total_marks / num_marks
    
    # Print the result
    print(f"Average Marks: {average_marks:.1f}")

# Input: List of marks
marks = [85, 72, 91, 65, 78]

# Call the function with the list of marks
calculate_average(marks)
```
