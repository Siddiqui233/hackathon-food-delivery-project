<h1>Employee Performance Bonus Eligibility
</h1>

```python
# Input: Dictionary of employees and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find the highest performance score
max_score = max(employees.values())

# Step 2: Find all employees who have the highest score
top_performers = [name for name, score in employees.items() if score == max_score]

# Step 3: Display the result
top_performers_str = ", ".join(top_performers)  # Join names with a comma
print(f"Top Performers Eligible for Bonus: {top_performers_str} (Score: {max_score})")
```

<h1>Search Query Keyword Analysis</h1>

```python
import string

# Input: Search query sentence
query = "Buy mobile phone buy phone online"

# Step 1: Convert the input to lowercase
query_lower = query.lower()

# Step 2: Remove common punctuation
query_clean = query_lower.translate(str.maketrans('', '', string.punctuation))

# Step 3: Split the query into words
words = query_clean.split()

# Step 4: Count the frequency of each word
word_frequency = {}
for word in words:
    word_frequency[word] = word_frequency.get(word, 0) + 1

# Step 5: Display only keywords that appear more than once
result = {word: count for word, count in word_frequency.items() if count > 1}

# Output the result
print(result)
```

<h1>Sensor Data Validation
</h1>

```python
# Input: List of sensor readings where the index represents the hour
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Step 1: Identify even (valid) readings
valid_readings = []

# Step 2: Loop through the list, checking for even readings
for hour, reading in enumerate(sensor_readings):
    if reading % 2 == 0:  # Check if the reading is even
        valid_readings.append((hour, reading))  # Store the hour and value as a tuple

# Step 3: Display the valid sensor readings
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

<h1> Email Domain Usage Analysis
</h1>

```python
# Input: List of employee email IDs
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Extract the email domains
domains = [email.split('@')[1] for email in emails]  # Get domain part after '@'

# Step 2: Count the frequency of each domain
domain_count = {}
for domain in domains:
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Step 3: Calculate the percentage usage of each domain
total_emails = len(emails)
domain_percentage = {domain: (count / total_emails) * 100 for domain, count in domain_count.items()}

# Step 4: Display the result
for domain, percentage in domain_percentage.items():
    print(f"{domain}: {percentage:.0f}%")
```

<h1>: Sales Spike Detection</h1>

```python
# Input: List of daily sales values
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate the average daily sales
average_sales = sum(sales) / len(sales)

# Step 2: Detect days with sales more than 30% above average
spike_threshold = average_sales * 1.30  # 30% above average
spike_days = [(index + 1, sale) for index, sale in enumerate(sales) if sale > spike_threshold]

# Step 3: Display the day number and sale value
for day, sale in spike_days:
    print(f"Day {day}: {sale}")
```
