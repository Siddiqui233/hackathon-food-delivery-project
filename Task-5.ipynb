<h1>Social Media – Post Engagement Analyzer</h1>

```python
def analyze_engagement(likes_list):
    # Initialize a variable to store the total likes
    total_likes = 0
    
    # Loop through the list of likes and sum them up
    for likes in likes_list:
        total_likes += likes
    
    # Determine the post status based on total likes
    if total_likes >= 1000:
        post_status = "Viral Post"
    else:
        post_status = "Normal Engagement"
    
    # Print the results
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {post_status}")

# Example usage:
likes = [200, 150, 300, 100, 500]  # List of likes on the post (could be from different users or sources)
analyze_engagement(likes)
```

<h1>Agriculture – Rainfall Adequacy Checker
</h1>

```python
def check_rainfall(rainfall_data, required_level=70):
    # Calculate the total rainfall using a loop
    total_rainfall = 0
    for rainfall in rainfall_data:
        total_rainfall += rainfall
    
    # Calculate the average rainfall
    average_rainfall = total_rainfall / len(rainfall_data)
    
    # Check if the average rainfall is adequate
    if average_rainfall >= required_level:
        rainfall_status = "Adequate Rainfall"
    else:
        rainfall_status = "Inadequate Rainfall"
    
    # Print the results
    print(f"Average Rainfall: {average_rainfall:.2f}")
    print(f"Rainfall Status: {rainfall_status}")

# Example usage:
rainfall_data = [85, 72, 60, 90, 75, 80]  # List of rainfall data for the season (in mm)
check_rainfall(rainfall_data)
```

<h1>Social Media – Duplicate Account Detection
</h1>

```python
def detect_duplicates(usernames):
    # Create an empty set to track unique usernames
    seen_usernames = set()
    
    # Loop through the list of usernames
    for username in usernames:
        # If the username is already in the set, it's a duplicate
        if username in seen_usernames:
            return "Duplicate Accounts Found: Yes"
        # Otherwise, add the username to the set
        seen_usernames.add(username)
    
    # If no duplicates are found, return unique accounts message
    return "Duplicate Accounts Found: No"

# Example usage:
usernames = ["Majeed", "Ahmed", "Siddiqui", "Iran", "Majeed"]  # List of usernames
result = detect_duplicates(usernames)
print(result)
```

<h1>Agriculture – Premium Crop Price Filter</h1>

```python
def filter_premium_crops(prices):
    # Filter out the prices that are above ₹2000
    premium_crops = [price for price in prices if price > 2000]
    
    # Print the result
    print(f"Premium Crops: {premium_crops}")

# Example usage:
prices = [1500, 2500, 3200, 1800, 2200]  # List of crop prices
filter_premium_crops(prices)
```

<h1>System Monitoring – Application Health</h1>

```python
def check_application_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    
    print(f"Error Count: {errors}")
    print(f"System Status: {status}")

# Example usage
check_application_health(7)
```

<h1> E-Learning – Student Attendance Eligibility System
</h1>

```python
def check_exam_eligibility(attendance_list):
    total_classes = len(attendance_list)
    attended_classes = 0

    for status in attendance_list:
        if status == 1:   # 1 means Present
            attended_classes += 1

    attendance_percentage = (attended_classes / total_classes) * 100

    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print(f"Attendance Percentage: {attendance_percentage}")
    print(f"Exam Eligibility: {eligibility}")


# Example usage
# 1 = Present, 0 = Absent
student_attendance = [1, 1, 1, 1, 0]  # 4 out of 5 classes attended (80%)
check_exam_eligibility(student_attendance)
```
