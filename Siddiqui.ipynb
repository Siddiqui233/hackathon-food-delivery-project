<h2>  Pass / Fail Analyzer  </h2>

```python
# Given marks of students
marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_count = 0
fail_count = 0

# Loop through the marks to count pass and fail students
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print the final result
print(f"Total Pass Students: {pass_count}")
print(f"Total Fail Students: {fail_count}")
```

<h1> 3. Simple Data Cleaner </h2>

```python
# Given list of names
names = [" Alice ", "bob", " CHARLIE "]

# Clean the names: strip spaces and convert to lowercase
cleaned_names = [name.strip().lower() for name in names]

# Print the cleaned list
print("Cleaned Names:", cleaned_names)
```

<h1>  Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
messages
```

```python
# Loop through the messages and analyze their lengths
for message in messages:
    message_length = len(message)  # Get the length of the message
    print(f"Message: '{message}' - Length: {message_length}")
```

```python
 # If the message is longer than 10 characters, flag it
if message_length > 10:
print("** Flagged: This message is longer than 10 characters! **")
```
